Creating clusters of the songs

```python
# 1. import the model
from sklearn.cluster import KMeans
 
# 2. initialize the model
my_kmeans = KMeans(n_clusters= 4)
 
# 3. fit the model to the data
my_kmeans.fit(df_minmax) # pass your scaled data here
 
# 4. obtain the cluster output
clusters = my_kmeans.predict(df_minmax) # pass your scaled data here
 
# 5. attach the cluster output to our original DataFrame
df["cluster"] = clusters
df
```

```python
df.groupby(by="cluster").mean()
```

```python
centroids = my_kmeans.cluster_centers_
df_centroids = pd.DataFrame(centroids)
df_centroids.columns = ["danceability", "energy", "key", "loudness", "mode", "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "duration_ms", "time_signature"]
df_centroids
```

```python
def euclidean_distance(data, cluster, cluster_to_compare): 
    distance = (
        pd.DataFrame({
            "cluster_1":data.loc[cluster,:],
            "cluster_2":data.loc[cluster_to_compare,:]
        })
        # applying the formula
        .assign(dist_abs = lambda x: (x.cluster_1 - x.cluster_2)**2)
        .agg({"dist_abs":"sum"})[0]**0.5
    )
    return distance

distances = [[euclidean_distance(df_centroids, cluster, cluster_to_compare) for cluster in df_centroids.index] for cluster_to_compare in df_centroids.index]
eucl_dist = pd.DataFrame(distances, columns=df_centroids.index, index=df_centroids.index)
plt.subplots(figsize=(30, 20))
sns.heatmap(eucl_dist)
```

```python
# plotting danceability through the clusters
fig, ax = plt.subplots(figsize=(30, 20))
sns.barplot(data=df_centroids, x=df_centroids.index, y="danceability")
plt.title("Danceability in each cluster")
plt.show()
```

```python
# scatterplot danceability and energy
fig, ax = plt.subplots(figsize=(20, 10))
sns.scatterplot(data=df_centroids, x="energy", y="danceability", hue=df_centroids.index)
plt.title("Danceability in each cluster")
plt.show()
```

Inertia score for the scale data "df_minmax"

```python
from sklearn.cluster import KMeans
myKMeans = KMeans(n_clusters=2)
myKMeans.fit(df_minmax)
myKMeans.inertia_
```

```python
inertia_list = []
for i in range(1,30):
    myKMeans = KMeans(n_clusters=i)
    myKMeans.fit(df_minmax)
    inertia_list.append(round(myKMeans.inertia_))
inertia_list
```

Visualization of the inertia for x clusters to check the correct amount of clusters --> "Elbow method"

Inertia = Distance of each instance to the centroid of the cluster

```python
max_k = 15
inertia = []

for k in range(1, max_k): 
    inertia.append(KMeans(n_clusters=k).fit(df_minmax).inertia_)

fig, ax = plt.subplots(figsize=(12, 6))
plt.title(f"Inertia evolution from 1 cluster to {max_k} cluster")
sns.lineplot(x=range(1, max_k), y=inertia, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

Silhouette score

```python
from sklearn.metrics import silhouette_score
silhouette_score(df_minmax, my_kmeans.labels_)
```

The silhouette coefficient can vary between -1 and +1: 

|   |   |
|---|---|
|+1   | the instance is well inside its own cluster and far from other clusters.  |
|0   | the instance is close to a cluster boundary.   |
|-1   | the instance may have been assigned to the wrong cluster.  |
